```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
train=pd.read_csv('train.csv')
print(train["SalePrice"].describe())
## to get some idea regarding the distribution of the target variable
train["SalePrice"].skew()
plt.hist(train["SalePrice"], bins=50)
## since data is right skewed, we will apply log transformation to make it resemble a normal distribution more
train["SalePrice"].skew()
```

```python
plt.hist(np.log1p(train["SalePrice"]), bins=50)
```

```python
## Thus we see that log transformation of the distribution of Sale Prices is more closely ressembling a normal distribution 
```
